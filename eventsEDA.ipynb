```python
file1 = open(file_name, 'r') 
count = 0
  
while count<1000: 
    count += 1
  
    # Get next line from file 
    line = file1.readline() 
  
    # if line is empty 
    # end of file is reached 
    if not line: 
        break
    print("Line{}: {}".format(count, line.strip())) 
  
file1.close() 
```
